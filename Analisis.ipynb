```python
df_resultado = pd.read_csv("./resultados/resultado.csv")
df_resultado = df_resultado[df_resultado["Método"] == 1]
df_resultado = df_resultado[(df_resultado["Vecinos"] == 7)]
df_resultado = df_resultado[(df_resultado["Alfa"] == 26)]
df_resultado = df_resultado[(df_resultado["Numero de iteraciones"] == 200)]

df_5000 = df_resultado[(df_resultado["Resultados"].str.find('5000', 0 ,22) != -1)]
DataSetSize = [5000 for i in range(0,df_5000["Alfa"].size)]
df_5000["DataSetSize"] = DataSetSize

df_10000 = df_resultado[(df_resultado["Resultados"].str.find('10000', 0 ,24) != -1)]
DataSetSize = [10000 for i in range(0,df_10000["Alfa"].size)]
df_10000["DataSetSize"] = DataSetSize

df_15000 = df_resultado[(df_resultado["Resultados"].str.find('15000', 0 ,24) != -1)]
DataSetSize = [15000 for i in range(0,df_15000["Alfa"].size)]
df_15000["DataSetSize"] = DataSetSize

df_20000 = df_resultado[(df_resultado["Resultados"].str.find('20000', 0 ,24) != -1)]
DataSetSize = [20000 for i in range(0,df_20000["Alfa"].size)]
df_20000["DataSetSize"] = DataSetSize

df_25000 = df_resultado[(df_resultado["Resultados"].str.find('25000', 0 ,24) != -1)]
DataSetSize = [25000 for i in range(0,df_25000["Alfa"].size)]
df_25000["DataSetSize"] = DataSetSize

df_30000 = df_resultado[(df_resultado["Resultados"].str.find('30000', 0 ,24) != -1)]
DataSetSize = [30000 for i in range(0,df_30000["Alfa"].size)]
df_30000["DataSetSize"] = DataSetSize

df_35000 = df_resultado[(df_resultado["Resultados"].str.find('35000', 0 ,24) != -1)]
DataSetSize = [35000 for i in range(0,df_35000["Alfa"].size)]
df_35000["DataSetSize"] = DataSetSize

df_40000 = df_resultado[(df_resultado["Resultados"].str.find('40000', 0 ,24) != -1)]
DataSetSize = [40000 for i in range(0,df_40000["Alfa"].size)]
df_40000["DataSetSize"] = DataSetSize

df_42000 = df_resultado[(df_resultado["Resultados"].str.find('7_fold', 0 ,20) != -1)]
DataSetSize = [42000 for i in range(0,df_42000["Alfa"].size)]
df_42000["DataSetSize"] = DataSetSize

df_final = pd.concat([df_5000, df_10000, df_15000, df_20000, df_25000, df_30000, df_35000, df_40000, df_42000])

fig = sns.lineplot(data=df_final, x='DataSetSize', y='Accuracy', color="purple");
fig.set(xlabel='DataSetSize', ylabel='Accuracy');
handles, labels = fig.get_legend_handles_labels()
plt.savefig("./imagenes/accuracy-knn+pca-DataSetSize_5000-40000_de_a_5000.svg");
```
